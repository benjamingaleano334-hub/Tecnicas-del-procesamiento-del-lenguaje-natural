# Modelo de Espacio Vectorial

-------------------------------
### Importar librerías


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```


-------------------------------
### Documentos


```python

docs = {
    "doc1": "El veloz zorro marrón salta sobre el perro perezoso.",
    "doc2": "Un perro marrón persiguió al zorro.",
    "doc3": "El perro es perezoso."
}
```



-------------------------------
# Vectorización con TF-IDF


```python

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs.values())
```


-------------------------------
# Calcular similitud del coseno

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```


-------------------------------
# Mostrar matriz de similitud


```python
print("Matriz de similitud del coseno:")
print(cosine_sim)
```

-------------------------------
# Graficar matriz de similitud


```python
plt.figure(figsize=(6,5))
sns.heatmap(cosine_sim, annot=True, cmap="Blues", 
            xticklabels=docs.keys(), yticklabels=docs.keys())
plt.title("Matriz de Similitud del Coseno")
plt.show()
```
